```python
import os
import numpy as np
import pandas as pd
from activity_classifier import ChunkableTimeSeries, CAP24_DIR
import activity_classifier as ac

FP_GAP_ANALYSIS_OUT = '/Users/bryanfry/code/activity-classifier/activity_classifier/EDA_output/EDA_01_check_gaps.csv'
```

```python
def get_raw_data_stats(df, file_path):
    """
    Return basic stats on raw data file as a pd.Series
    """
    fn = os.path.basename(file_path)
    n = len(df)
    dur = df.iloc[-1]['timestamp'] - df.iloc[0]['timestamp']
    diff = np.diff(df['timestamp'].values)
    max_diff = np.max(diff)
    min_diff = np.min(diff)
    vc = pd.value_counts(diff)
    return pd.Series({
        'file_name': fn,
        'n_samples': n,
        'duration': dur,
        'max_diff': max_diff,
        'min_diff': min_diff,
        'diff_distribution': vc
        })


FPS_GZ_DATA = sorted([f'{CAP24_DIR}/{f}' for f in os.listdir(CAP24_DIR) if f.endswith(".csv.gz")])

ss = []  # To populate with stats Series elements
for i, fp in enumerate(FPS_GZ_DATA):
    print('processing file: ', os.path.basename(fp), f'({i+1}/{len(FPS_GZ_DATA)})')
    df = ac.load_acc_data_file(fp, convert_datetime=True, add_timestamp_col=True)
    stats = get_raw_data_stats(df, fp)
    #print(stats)
    ss.append(stats)
    print('DONE')

dfg = pd.DataFrame(ss)
dfg.to_csv(FP_GAP_ANALYSIS_OUT, index=False)
```
